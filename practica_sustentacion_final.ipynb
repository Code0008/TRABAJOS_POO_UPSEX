```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_students_survey = pd.read_csv("StudentSurvey.csv")
data_students_survey
```

```python
data_particionada_por_genero = data_students_survey.groupby(by="Gender").sum().reset_index()
piercings_por_sexo = data_particionada_por_genero.loc[:, ["Gender","Piercings"]]
plt.title("Piercings por sexo :v")
plt.pie(piercings_por_sexo["Piercings"], explode=(0.01,)*2, autopct="%1.1f%%", labels=piercings_por_sexo["Gender"]);
```

```python
agrupacion = data_students_survey.groupby(by=["Year", "Smoke","Gender"]).size()

plt.figure(figsize=(12,12))
plt.pie(agrupacion, labels=agrupacion.index, explode=(0.01,)*len(agrupacion), autopct="%1.1f%%");
```

```python
cultivos = pd.read_csv("cultivos_regiones.csv")
cultivos
```

```python
agrupando_por_año_region = cultivos.groupby(by=["Año", "Región", "Cultivo"]).sum()
datoss_2015 = agrupando_por_año_region.loc[(2015)]
datoss_2015
```

```python
#Porcentaje de area cosechada en region y cultivo en el año 2015
plt.title("Porcentaje de area cosechada en region \n y cultivo en el año 2015")
plt.pie(datoss_2015["Área Cosechada (ha)"], explode=(0.01,)*len(datoss_2015), autopct="%1.1f%%", labels=datoss_2015.index)
```
