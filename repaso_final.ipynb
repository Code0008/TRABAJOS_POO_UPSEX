```python


def contador(caracter):
    matrez = np.random.randint(ord('A'), ord('Z'), size=(10,10))
    datos_matrez_tratar = matrez.flatten()
    return datos_matrez_tratar.tolist().count(ord(caracter))

def cantidad_repeticiones(veces_repetir, caracter):
    resultados = []
    for _ in range(veces_repetir):
        resultados.append(contador(caracter=caracter))
    return resultados

def dispercion_resultados(arreglo: np.array):
    return f"""
    Variacion: {arreglo.var()}
    Desviacion estandar: {arreglo.std()}
    Rango: {arreglo.max()- arreglo.min()}

        """
def medidas_tendencia(arreglo):
    return f"""

        Mediana: {arreglo.mean()}
        """

caracter = "B"
resultados = cantidad_repeticiones(1000, caracter=caracter)
resultados = np.array(resultados)
print(resultados)
print(dispercion_resultados(resultados))
print(medidas_tendencia(resultados))
plt.hist(resultados)
plt.figure(figsize=(10,10))
plt.scatter(np.arange(0,1000), resultados)
plt.ylim(0,20)

```

```python
class Terreno:
  __diccionario = {}
  def __init__(self,codigo,ubicacion,area,valor_m2_comercial,valorm2predial):
    self.__codigo = codigo
    self.__ubicacion = ubicacion
    self.__area = area
    self.__valor_m2_comercial = valor_m2_comercial
    self.__valorm2predial = valorm2predial
  #get
  def getCodigo(self):
    return self.__codigo
  def getUvbacion(self):
    return self.__ubicacion
  def getArea(self):
    return self.__area
  def valor_m2_comercial(self):
    return self.__valor_m2_comercial
  def valor_m2_predial(self):
    return self.__valorm2predial
  #set
  def setCodigo(self,codigo):
    self.__codigo = codigo
  def setUbicacion(self,ubicacion):
    self.__ubicacion = ubicacion
  def setArea(self,area):
    self.__area = area
  def setValor_m2_comercial(self,valor_m2_comercial):
    self.__valor_m2_comercial = valor_m2_comercial
  def setValor_m2_predial(self,valorm2predial):
    self.__valorm2predial = valorm2predial

  def valor_predial_terreno(self):
    return self.__area * self.__valorm2predial
  def valor_comercial_terreno(self):
    return self.__area * self.__valor_m2_comercial
  def margen_ganancia(self):
    return self.valor_comercial_terreno() - self.valor_predial_terreno()

  def agregar(self):
    self.__diccionario[self.__codigo] = self.__ubicacion,self.__area,self.valor_comercial_terreno(),self.valor_predial_terreno(),self.margen_ganancia()
  def listar(self):
    for clave,valor in self.__diccionario.items():
      print(f'{clave}: {valor}')


obj_terreno = Terreno("asdf","28 de Julio",1000,50000,1000)
obj_terreno.agregar()
obj_terreno.listar()
obj_terreno = Terreno("werr","3 de mayo",2000,70000,4000)
obj_terreno.agregar()
obj_terreno.listar()
```

```python
import pandas as pd
import numpy as np
from random import choice, randint
class CandidatoGerrero:
    def __init__(self) -> None:
        self.__notas = np.array([
                randint(10,100) for _ in range(4)
        ])
        self.__genero = choice(['F', "M"])
        self.cualidades_de_guerrero()

    @property
    def genero(self):
        return self.__genero
    @property
    def notas(self):
        return self.__notas
    @property
    def dd(self):
        return self.__estadisticas_evaluacion_desempe簽o
    def mostrar_notas_detalle(self):
        return  f"""
                  "Genero": {self.genero}
                "Teoricas", {self.notas[0]}
                "Resistencia", {self.notas[1]}
                "Combatecac",{self.notas[2]}
                "Estmil"{self.notas[3]}
                """

    
    def calcular_estadisticas_notas(self):
        return [
             self.notas.mean(),# promedio notas 
             self.notas.std()
        ]
    def cualidades_de_guerrero(self):
        habilidades_militares = [
    "Dominio de estrategia militar",
    "Combate cuerpo a cuerpo avanzado",
    "Manejo de armas de fuego",
    "T獺cticas de infiltraci籀n y sigilo",
    "Resistencia f穩sica y mental",
    "Capacidad de liderazgo en combate",
    "Toma de decisiones bajo presi籀n",
    "Conocimientos de primeros auxilios",
    "Capacitaci籀n en supervivencia",
    "Habilidad para operar equipo t獺ctico",
    "Trabajo en equipo efectivo",
    "Adaptabilidad a condiciones extremas",
    "Reconocimiento y an獺lisis del terreno",
    "Capacidad de comunicaci籀n en el campo de batalla"
  ]
        habilidades_puntajes = {valor: randint(50,100) for valor in habilidades_militares}
        arreglo = np.array(list(habilidades_puntajes.values()))
        self.__estadisticas_evaluacion_desempe簽o = [
            arreglo.std(),#std
            arreglo.mean()#mean
        ]
    

class GrupoCandidatos:
    def __init__(self,titan_candidato) -> None:
        self.__candidatos = [CandidatoGerrero() for _ in range(4)]
        self.__dataframe_candidatos = pd.DataFrame()
        self.__tipo = titan_candidato
    @property
    def tipo(self):
        return self.__tipo
    def obtener_datos_grupo(self):
        dataframe = pd.DataFrame(index=np.arange(0, len(self.__candidatos)), columns=[
                                                "Genero",
                                                "Teoricas", 
                                                "Resistencia", 
                                                "Combatecac",
                                                "Estmil", 
                                                "variacion_std" ,
                                                  "promedio_notas",
                                                  "tipo_titan",
                                                  "promedio_dd", 
                                                  "variacion_dd"
                                                    ])
        for indice,candidato in enumerate(self.__candidatos):
            varloes_estadisticos =  candidato.calcular_estadisticas_notas()
            valores_desempe簽o =  candidato.dd
            dataframe.iloc[indice,:] = (
                                        candidato.genero, 
                                        candidato.notas[0],
                                        candidato.notas[1],
                                        candidato.notas[2],
                                        candidato.notas[3],
                                        varloes_estadisticos[1],
                                        varloes_estadisticos[0], 
                                        self.tipo,
                                        valores_desempe簽o[1],
                                        valores_desempe簽o[0]
                                        )
        self.__dataframe_candidatos = dataframe
        return self.__dataframe_candidatos

            
            
    def acceder_candidato(self, id_busqueda):
        for candidato in self.__candidatos:
            if candidato._id == id_busqueda:
               candidato.mostrar_notas_detalle()
               break
    
class ManejadorCandidatos:
    def __init__(self) -> None:
        self.titanes = ["Colosal", "Cargero", "Acorazado", "Hembra","Bestia"]
        self.__grupos_candidatos = {titan:[GrupoCandidatos(titan) for _ in range(4)] for titan in self.titanes}

    def dataframe_general(self):
        self.__dataframes_generales_titanes = pd.DataFrame(columns=[  
                                                "Genero",
                                                "Teoricas",
                                                "Resistencia", 
                                                "Combatecac",
                                                "Estmil",
                                                "variacion_std" ,
                                                "promedio_notas",
                                                "promedio_dd", 
                                                "variacion_dd",
                                                "tipo_titan"
                                                ])
        for titan in self.titanes:
            for grupo  in self.__grupos_candidatos[titan]:
                self.__dataframes_generales_titanes = pd.concat([ self.__dataframes_generales_titanes,  grupo.obtener_datos_grupo()], ignore_index=True)
                 
        return self.__dataframes_generales_titanes
    
    def requisitos_colosal(self):
        quienes_superen_minimas = self.__dataframes_generales_titanes.where(((self.__dataframes_generales_titanes["Teoricas"]>=90)  & (self.__dataframes_generales_titanes["Resistencia"]>=80)) &(self.__dataframes_generales_titanes["tipo_titan"] == "Colosal") )
        quienes_superen_minimas.dropna(inplace=True)        
        if quienes_superen_minimas.empty:
            maximo = self.__dataframes_generales_titanes[self.__dataframes_generales_titanes["tipo_titan"]=="Colosal"]["promedio_notas"].max()
            return  self.__dataframes_generales_titanes.where((self.__dataframes_generales_titanes["tipo_titan"]=="Colosal") & (self.__dataframes_generales_titanes["promedio_notas"]==maximo)).dropna()
        return quienes_superen_minimas
    def requisitos_hembra(self):
        solo_mujeres = self.__dataframes_generales_titanes[(self.__dataframes_generales_titanes["Genero"]=='F') & (self.__dataframes_generales_titanes["tipo_titan"]=="Hembra")]
        maximo_promedio = solo_mujeres[(solo_mujeres["Resistencia"]>=85) ]
        if len(maximo_promedio)>1:
                return solo_mujeres[(solo_mujeres["Combatecac"]==solo_mujeres["Combatecac"].max() ) ]
        return maximo_promedio

    def requisitos_cargero(self):
        cargeros = self.__dataframes_generales_titanes[self.__dataframes_generales_titanes["tipo_titan"] == 'Cargero']
        cumplen_condiciones = cargeros[(cargeros["promedio_notas"]>=cargeros["promedio_notas"].max()) & (cargeros["Resistencia"]>=90)]
        if cumplen_condiciones.empty:
            return cargeros[cargeros["promedio_notas"]>=cargeros["promedio_notas"].max()]
        return cumplen_condiciones
    def reporte2(self):
        lista_filtrados = [ self.__dataframes_generales_titanes[self.__dataframes_generales_titanes["tipo_titan"]==titan] for titan in self.titanes ]

        for indice,filtrados in enumerate(lista_filtrados):        
            plt.bar(self.titanes[indice],filtrados.sort_values(by="promedio_notas").tail(len(filtrados)//10).reset_index(drop=True).loc[0,"promedio_notas"])
            #print(filtrados.sort_values(by="promedio_notas").head(len(filtrados)//10), indice)
    

man = ManejadorCandidatos()
man.dataframe_general()
man.reporte2()
```
