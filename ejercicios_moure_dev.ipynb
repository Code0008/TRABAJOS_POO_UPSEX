```python
def anillos_del_poder( **razas ):

    diccionario_bandos = {
        1:
        {"peleosos": 1, 
        "sure√±os buenos":2, 
        "enanos": 3,
        "numenoreanos": 4,
        "elfos": 5},
        2:{
        "sure√±os buenos": 2, 
        "orcos":2, 
        "goblins": 2,
        "huargos": 3,
        "trolls": 5
            }
    }

    def calcular_resultados_puntos(integrante, cantidad, ejercito):
        return diccionario_bandos[ejercito][integrante]*cantidad

    resultados = [
        sum([calcular_resultados_puntos(integrante, razas[integrante], 1) for integrante in diccionario_bandos[1]  if integrante in razas and integrante in diccionario_bandos[1]]),
        sum([calcular_resultados_puntos(integrante, razas[integrante], 2) for integrante in diccionario_bandos[2]  if integrante in razas and integrante in diccionario_bandos[2]])
    ]   
    return resultados
anillos_del_poder(elfos = 4, trolls=3, orcos=20)
```

```python
def triangulo_pascal(tamano):
    triangulo = []

    for indice in range(tamano):
        fila = [1]*(indice+1)
        for columna in range(1, indice):
            fila[columna]  = triangulo[indice-1][columna-1] + triangulo[indice-1][columna]
        triangulo.append(fila)
    return triangulo
triangulo_pascal(10)
```


 Este es un reto especial por Halloween.
 Deberemos crear un programa al que le indiquemos si queremos realizar "Truco
 o Trato" y un listado (array) de personas con las siguientes propiedades:
  - Nombre de la ni√±a o ni√±o
  - Edad
  - Altura en cent√≠metros
 
 Si las personas han pedido truco, el programa retornar√° sustos (aleatorios)
 siguiendo estos criterios:
  - Un susto por cada 2 letras del nombre por persona
  - Dos sustos por cada edad que sea un n√∫mero par
  - Tres sustos por cada 100 cm de altura entre todas las personas
  - Sustos: üéÉ üëª üíÄ üï∑ üï∏ ü¶á
 
 Si las personas han pedido trato, el programa retornar√° dulces (aleatorios)
 siguiendo estos criterios:
  - Un dulce por cada letra de nombre
  - Un dulce por cada 3 a√±os cumplidos hasta un m√°ximo de 10 a√±os por persona
  - Dos dulces por cada 50 cm de altura hasta un m√°ximo de 150 cm por persona
  - Dulces: üç∞ üç¨ üç° üç≠ üç™ üç´ üßÅ üç©
  - En caso contrario retornar√° un error.


```python
class persona:
    def __init__(self, nombre,edad, altura, seleccion) -> None:
        self.__nombre = nombre
        self.__edad = edad
        self.__altura = altura
        self.__seleccion =  seleccion
    @property
    def nombre(self):
        return self.__nombre
    @property
    def edad(self):
        return self.__edad
    @property
    def altura(self):
        return self.__altura
    @property
    def seleccion(self):
        return self.__seleccion
    
class jalowin:
    def __init__(self) -> None:
        self.__personas = [persona(
            nombre=''.join([choice(list(ascii_letters)) for _ in range(randint(3, 10))]),
            edad=randint(3, 100),
            altura=randint(140, 200),
            seleccion=randint(1,2)) for _ in range(100)]
    @property
    def personas(self):
        return self.__personas
    def calcular_resultados(self):
        def truco(persona):
            return sum([
                len(persona.nombre)//2,
                2 if persona.edad % 2 == 0 else 0,
                3*(persona.altura//100)
            ])
        def trato(persona):
            return sum([
                len(persona.nombre),
               persona.edad//3 if persona.edad <= 10 else 0,
               2 if persona.altura < 150 else 3
            ])  

        resultados= {
            "trucos": [],
            "tratos": []
        }
        for persona in self.__personas:
            match persona.seleccion:
                case 1: resultados["tratos"].append(trato(persona))
                case 2: resultados["trucos"].append(truco(persona))
        return resultados
    
manejador = jalowin()
manejador.calcular_resultados()
```

```python
def calcular_agua_trapada(bloques):
    if len(bloques) < 3:
        # No puede haber agua atrapada si hay menos de 3 bloques
        return 0

    # Inicializaci√≥n de punteros y variables
    izquierda = 0
    derecha = len(bloques) - 1
    max_izq = bloques[izquierda]
    max_der = bloques[derecha]
    agua_trapada = 0

    # Recorremos el array con dos punteros
    while izquierda < derecha:
        if max_izq < max_der:
            izquierda += 1
            max_izq = max(max_izq, bloques[izquierda])
            agua_trapada += max(0, max_izq - bloques[izquierda])
        else:
            derecha -= 1
            max_der = max(max_der, bloques[derecha])
            agua_trapada += max(0, max_der - bloques[derecha])

    return agua_trapada

# Ejemplo
bloques = [4, 0, 3, 6, 1, 3]
agua = calcular_agua_trapada(bloques)
print(f"Unidades de agua atrapadas: {agua}")
```


  Calcula d√≥nde estar√° un robot (sus coordenadas finales) que se
  encuentra en una cuadr√≠cula representada por los ejes "x" e "y".
  - El robot comienza en la coordenada (0, 0).
  - Para idicarle que se mueva, le enviamos un array formado por enteros
    (positivos o negativos) que indican la secuencia de pasos a dar.
  - Por ejemplo: [10, 5, -2] indica que primero se mueve 10 pasos, se detiene,
    luego 5, se detiene, y finalmente 2.
    El resultado en este caso ser√≠a (x: -5, y: 12)
  - Si el n√∫mero de pasos es negativo, se desplazar√≠a en sentido contrario al
    que est√° mirando.
  - Los primeros pasos los hace en el eje "y". Interpretamos que est√° mirando
    hacia la parte positiva del eje "y".
  - El robot tiene un fallo en su programaci√≥n: cada vez que finaliza una
    secuencia de pasos gira 90 grados en el sentido contrario a las agujas
    del reloj.
 

```python
def donde_esta_el_robot(pasos):
    cuadrantes = [90, 180, 270, 360]
    suma_eje_y = 0
    suma_eje_x = 0
    # inicia en 0,0
    # pasos = [10, 5, -2]
    pasitos = 0
    for indice in range(len(pasos)):
        if pasitos==len(cuadrantes):
            pasitos=0
        cuadrante_actual = cuadrantes[pasitos]
        pasitos+=1

        match cuadrante_actual:
            case 90:
                suma_eje_y+= pasos[indice]
            case 180:
                suma_eje_x-= pasos[indice]
            case 270:
                suma_eje_y-= pasos[indice]
            case 360:
                suma_eje_x+= pasos[indice]
        print(suma_eje_x, suma_eje_y)

    return suma_eje_x, suma_eje_y
donde_esta_el_robot([10, 5, -2, 4, 8, 3,12,23123,334,123])     

```

```python
class Pokedex:
    def __init__(self) -> None:
        self.__pokemones_consultados = {}

    def consumir_api(self, pokemon):
       self.__pokemones_consultados[pokemon] = requests.get(f"https://pokeapi.co/api/v2/pokemon/{pokemon}").json()

    def ver_habilidades(self, pokemon):
        habilidades = self.__pokemones_consultados[pokemon]["abilities"]
        habilidades = [habilidad["ability"]["name"] for habilidad in habilidades]
        return habilidades
    def ver_movimientos(self, pokemon):
        movimientos =  self.__pokemones_consultados[pokemon]["moves"]
        movimientos = [movimiento["move"]["name"] for movimiento in movimientos]
        return movimientos
poke = Pokedex()
poke.consumir_api("pikachu")
poke.ver_habilidades("pikachu")
poke.ver_movimientos("pikachu")
```
