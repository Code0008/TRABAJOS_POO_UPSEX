```python
class Cuenta_Bancaria:
    def __init__(self, numero_cuenta = 0, titular= "", saldo="") -> None:
        self.__numero_cuenta = numero_cuenta
        self.__titular = titular
        self.__saldo = saldo
        
    def get_titular(self):
        return self.__titular
    def get_saldo(self):
        return self.__saldo
    def deposito(self, saldo):
        self.__saldo += saldo
    def retiro(self, monto):
        self.__saldo -= monto

    def __str__(self) -> str:
        return f"""
    Numero Cuenta: {self.__numero_cuenta}
    Saldo: {self.__saldo}
    """


class Persona:
    def __init__(self, nombre) -> None:
        self.__nombre = nombre
        self.__cuentas_bancarias = []

    def get_nombre(self):
        return self.__nombre
    def get_cuentas_persona(self):
        return self.__cuentas_bancarias
    def agregar_cuenta_bancaria(self, numero_cuenta, saldo_inicial):
        self.__cuentas_bancarias.append(Cuenta_Bancaria(numero_cuenta=numero_cuenta, 
                                                        saldo=saldo_inicial,
                                                        titular=self.get_nombre()
                                                        ))
    def mostrar_cuentas(self):
        for cuenta in self.__cuentas_bancarias:
            print(cuenta)


class Banco:
    __nombre_banco ="Banco Tilinesco"
    def __init__(self) -> None:
        self.__personas_afiliadas = []
        #Nombre:  [cuentas]

    def agregar_personas(self):
        nombre = ''.join([choice([chr(c) for c in range(ord('A'), ord('Z'))]) for _ in range(0, 8)])
        persona =  Persona(nombre=nombre)
        persona.agregar_cuenta_bancaria(numero_cuenta=''.join([str(randint(0,9)) for _ in range(20)]),
                                        saldo_inicial= round(randint(1,200000) / randint(0, 100), 2)
                                        )
        self.__personas_afiliadas.append(persona)
    
    def mostrar_todas_las_cuentas(self):
        for persona in self.__personas_afiliadas:
            current_cuentas= persona.get_cuentas_persona()
            print(f"""
            Titular: {persona.get_nombre()}
            Tiene {len(current_cuentas)}
            """)
            persona.mostrar_cuentas()
    def mostrar_cuenta_de_persona_especifica(self, nombre):
        for persona in self.__personas_afiliadas:
            if persona.get_nombre() == nombre:
                persona.mostrar_cuentas()
                break
    

banco = Banco()
banco.agregar_personas()
banco.agregar_personas()

banco.mostrar_todas_las_cuentas()





```

### ***Caso 2***
- La Administrador de la entidad financiera “FinanTek” requiere de un programa orientado a objetos para otorgar créditos para familias emprendedoras en Lima.
- Cada representante familiar deberá inscribirse con los siguientes datos: DNI, edad, número de hijos menores a 18 años, ingreso familiar mensual (en soles) y m2 de propiedad (el total de metros cuadrados de los que el miembro es propietario, si no tiene bienes el valor será cero).
- Con estos datos se generará una calificación, cuyo valor se calcula así:
  - calificación = edad + número de hijos + (ingreso familiar mensual) / (m2 de propiedad +1)
- Según la calificación se le puede otorgar un crédito con los siguientes rangos:

|Calificación|Monto Máximo (S/.)|
|---|----|
|0<150|35,000.00|
|151-350|45,000.00|
|>350|60,000.00|

1. Ingrese datos de representantes.
2. Genere un listado de todos los representantes (incluya la calificación obtenida por cada representante).
3. Dado un monto de crédito muestre que representantes logran alcanzar dicho monto.
4. Desarrolle un método que permita obtener monto total que se desembolsaría para lo S representantes que tengan más de 2 hijos.
5. Desarrolle un método que permita buscar un representante familiar por DNI, y que devuelva los datos del representante, así como el crédito que se le puede dar. En caso no se encuentre el miembro debe indicar “No encontrado”.
6. Genere un menú que permita administrar lo descrito anteriormente


```python
#jija caso 2

class validadores():
    def validar_dni(self):
        dni= ""
        while not (len(dni)==8 and all([True for e in dni if e.isdigit() ])):
            dni = input("dni:")
        return dni
    def validar_rango_numero_positivo_entero(self, limite_a, limite_b, mensaje):
        ingreso = -1
        while not (limite_a<ingreso<limite_b and type(ingreso)== int):
            ingreso= int(input(f"Ingrese {mensaje} "))
        return ingreso
    def validar_rango_numero_positivo_flotante(self, limite_a, limite_b, mensaje):
        ingreso = -1
        while not (limite_a<ingreso<limite_b and type(ingreso)== float ):
            ingreso= float(input(f"Ingrese {mensaje} "))
        return ingreso
class Representante:
    def __init__(self, dni, edad, numero_hijos, mensual_familiar, tamano_propiedad) -> None:
        self.__dni = dni
        self.__edad = edad
        self.__numero_hijos= numero_hijos
        self.__ingreso_mensual_familiar = mensual_familiar
        self.__tamano_propiedad = tamano_propiedad if tamano_propiedad>0 else 0
        self.__credito = self.asignar_credito()
    def get_numero_hijos(self):
        return self.__numero_hijos
    def get_credito(self):
        return self.__credito
    def get_dni(self):
        return self.__dni
    def calificacion(self):
        return self.__edad + self.__numero_hijos + (self.__ingreso_mensual_familiar/ (self.__tamano_propiedad  +1 ))
    def  asignar_credito(self):
        calificacion = self.calificacion()

        if 0<=calificacion<=150:
            return  35000.00
        if 151<=calificacion<=350:
            return 45000.00
        if calificacion>350:
            return 60000.00
        return 0
    def __str__(self) -> str:
        return f"""
    Representante: {self.__dni}
    edad: {self.__edad}
    Numero hijos: {self.__numero_hijos}
    Ingreso familiar : {self.__ingreso_mensual_familiar}
    Tamano propiedad: {"no tiene propiedad" if self.__tamano_propiedad == 0 else self.__tamano_propiedad}
    Credito asignado: {self.__credito}
    """
class FinanTek:
    def __init__(self) -> None:
        self.__lista_representantes = []


    def ingresar_representante(self):
        validador = validadores()
        self.__lista_representantes.append(Representante(
            dni=validador.validar_dni(),
            edad=validador.validar_rango_numero_positivo_entero(limite_a=18, limite_b=100, mensaje="su edad"),
            numero_hijos=validador.validar_rango_numero_positivo_entero(limite_a=-1, limite_b=100, mensaje="numero de hijos"),
            mensual_familiar=validador.validar_rango_numero_positivo_flotante(limite_a=100, limite_b=100000, mensaje="el monto familiar"),
            tamano_propiedad=validador.validar_rango_numero_positivo_flotante(limite_a=-1, limite_b=1000, mensaje="Tamano de su vivienda"),
        ))

    def mostrar_representantes(self):
        for representante in self.__lista_representantes:
            print(representante)
    def monto_total_mas_dos_hijos(self):
        monto_total = 0
        for representante in self.__lista_representantes:
            if representante.get_numero_hijos()>2:
                monto_total+=representante.get_credito()
    
    def haberiguar_que_representante_alcanza_monto(self, monto):
        for representante in self.__lista_representantes:
            if representante.get_credito()== monto:
                print(representante)
    def buscar_representante(self, dni_buscar):
        for representante in self.__lista_representantes:
            if representante.get_dni() == dni_buscar:
                print(representante)
                return 
        print(f"No encontrado")

finan = FinanTek()
finan.ingresar_representante()
finan.mostrar_representantes()
```
